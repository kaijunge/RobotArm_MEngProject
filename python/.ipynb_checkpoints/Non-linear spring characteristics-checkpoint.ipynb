```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to display something quickly
def plot_y(y, f=0, title = ""):
    plt.figure(f)
    x = np.linspace(0, len(y), len(y))
    plt.plot(x, y)
    if title != "":
        plt.title(title)
    plt.show
```

```python
# return theta value based on y
def find_theta(y, d, r1, r2):
    p = np.sqrt(y**2 + d**2)
    return np.arcsin((r1+r2)/p) + np.arctan(y/d)
```

```python
def find_rope_length(y, d, r1, r2):
    theta = find_theta(y, d, r1, r2)
    
    l1 = r1*theta
    l2 = r2*theta
    l3 = (r1 + r2)*(1/np.tan(theta)) + y/np.sin(theta)
    
    return (l1 + l2 + l3)*2, theta
```

```python
def plot_setup(d, r1, r2):
    
    h_feed = np.linspace(y_max + y_safety - (r1+r2), y_safety - (r1+r2), 1000)

    l, theta = find_rope_length(h_feed, d, r1, r2)
    
    max_length = l[0]
    for i in range (len(l)):
        l[i] = max_length - l[i]
    
    k = 1
    F_spring = k * np.linspace(0, y_max, 1000) #need to include spring constant
    F_rope   = F_spring / (2 * np.sin(theta))
    
    return l, h, F_rope
```

```python
# distance between circle centres in mm
d = 50

# radius of each pulley in mm. r1 => pulley with spring, r2 => rest pulley
r1 = 15
r2 = 15

# total movable distance in mm
y_max = 35
h = np.linspace(0, y_max, 1000)

# vertical distance to prevent F = âˆž
y_safety = 10
```

```python
fig, ax = plt.subplots(1,2, figsize = (15, 5))

for i in range(5):
    y_max = 30#30 + 5*i
    y_safety = 4 + i
    l, h, F_rope = plot_setup(30, 10, 5)

    p = 0
    ax[p].plot(l, h)
    ax[p].set_xlabel("L (rope extension)")
    ax[p].set_ylabel("H (spring contraction)")

    p = 1
    ax[p].plot(l, F_rope)
    ax[p].set_xlabel("L (rope extension)")
    ax[p].set_ylabel("F (resistive force)")
```
